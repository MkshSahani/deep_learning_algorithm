```python
import os
fnames = [os.path.join(cats_train_dir, f) for f in os.listdir(cats_train_dir)]
```

```python
img_file = fnames[10]
```

```python
import keras.utils as utils
```

```python
img = utils.load_img(img_file, target_size = (150, 150))
```

```python
arr = utils.img_to_array(img)
```

```python
arr.shape
```

```python
import matplotlib.pyplot as plt
brr = arr.copy()
```

```python
brr = brr.astype('uint8')
```

```python
plt.imshow(brr)
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
arr = arr.reshape((1,) + arr.shape)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 40,
    height_shift_range = 0.2,
    width_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    fill_mode = 'nearest'
)
```

```python
i = 0
for batch in datagen.flow(arr, batch_size = 1):
    plt.figure(i)
    imgplot = plt.imshow(utils.array_to_img(batch[0]))
    i += 1
    if i % 4 == 0:
        break
plt.show()
```
